```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python
import pandas as pd
data = pd.read_csv('data.csv')
data.head()
```

```python
dis_data = data.copy()
dis_data.drop('Highest_Qualified_Member', axis=1, inplace=True)
fig, axes = plt.subplots(ncols=len(dis_data.columns), figsize=(50,8))
for ax, col in zip(axes, dis_data.columns):
  sns.distplot(dis_data[col], ax=ax)
  plt.tight_layout() 
plt.show()
```

```python
mu_mthly_income = data['Mthly_HH_Income'].mean()
sigma_mthly_income = data['Mthly_HH_Income'].std()
print("mu_mthly_income", mu_mthly_income)
print("sigma_mthly_income", sigma_mthly_income)

mu_mthly_expense = data['Mthly_HH_Expense'].mean()
sigma_mthly_expense = data['Mthly_HH_Expense'].std()
print("\nmu_mthly_expense", mu_mthly_expense)
print("sigma_mthly_expense", sigma_mthly_expense)

mu_no_of_fly_members = data['No_of_Fly_Members'].mean()
sigma_no_of_fly_members = data['No_of_Fly_Members'].std()
print("\nmu_no_of_fly_members", mu_no_of_fly_members)
print("sigma_no_of_fly_members", sigma_no_of_fly_members)

mu_emi_or_rest_amt = data['Emi_or_Rent_Amt'].mean()
sigma_emi_or_rest_amt = data['Emi_or_Rent_Amt'].std()
print("\nmu_emi_or_rest_amt", mu_emi_or_rest_amt)
print("sigma_emi_or_rest_amt", sigma_emi_or_rest_amt)

mu_annual_income = data['Annual_HH_Income'].mean()
sigma_annual_income = data['Annual_HH_Income'].std()
print("\nmu_annual_income", mu_annual_income)
print("sigma_annual_income", sigma_annual_income)

mu_no_of_earning_members = data['No_of_Earning_Members'].mean()
mu_no_of_earning_members = data['No_of_Earning_Members'].std()
print("\nmu_no_of_earning_members", mu_no_of_earning_members)
print("mu_no_of_earning_members", mu_no_of_earning_members)
```

```python
one_std_right = mu_no_of_fly_members + (1 * sigma_no_of_fly_members)

one_std_left = mu_no_of_fly_members - (1 * sigma_no_of_fly_members)

two_std_right = mu_no_of_fly_members + (2 * sigma_no_of_fly_members)

two_std_left = mu_no_of_fly_members - (2 * sigma_no_of_fly_members)

three_std_right = mu_no_of_fly_members + (3 * sigma_no_of_fly_members)

three_std_left = mu_no_of_fly_members - (3 * sigma_no_of_fly_members)
```

```python
plt.figure(figsize=(20,10))
sns.set_style("darkgrid")
sns.distplot(data['No_of_Fly_Members'])

plt.axvline(mu_no_of_fly_members, color='coral', label='Mean')

plt.axvline(one_std_right, color='yellow', label='Mean + 1SD')
plt.axvline(one_std_left, color='yellow', label='Mean - 1SD')
plt.axvline(two_std_right, color='green', label='Mean + 2SD')
plt.axvline(two_std_left, color='green', label='Mean - 2SD')
plt.axvline(three_std_right, color='blue', label='Mean + 3SD')
plt.axvline(three_std_left, color='blue', label='Mean - 3SD')
plt.legend();
```
